```python
style.use("ggplot")

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    pullData = open("twitter-out.txt","r").read()
    lines = pullData.split('\n')

    xar = []
    yar = []

    x = 0
    y = 0

    for l in lines[0:]:
        x += 1
        if "pos" in l:
            y += 1
        elif "neg" in l:
            y -= 1
            
        xar.append(x)
        yar.append(y)
        
    ax1.clear()
    ax1.plot(xar,yar)
    plt.xlabel('Number of tweets')
    plt.ylabel('Tweets Sentiments (pos/neg)')
    plt.savefig('tweets_sentiment.png')
ani = animation.FuncAnimation(fig, animate, interval=100)
#plt.savefig('tweets_sentiment.png')
plt.show()
```
